a. Citit imaginile din fisier si returnate intr-un np.array

```python
import numpy as np
import os

file_path = r'C:\Users\anama\Documents\fmi\Anul II\Semestrul II\Inteligenta Artificiala\Machine Learning\Laborator\Laborator 1\images'
files = os.listdir(file_path)
images = []
for file in files:
    image_path = os.path.join(file_path, file)
    image = np.load(image_path)
    images.append(image)

images = np.array(images)

print(images)
```

b. Suma valorilor pixelilor tuturor imaginilor

```python
pixel_sum_total = np.sum(images)
print('pixel sum total =', pixel_sum_total)
```

c. Suma valorilor pixelilor pentru fiecare imagine

```python
pixel_sum = np.sum(images, axis=(1, 2))
print('pixel sum for each image =', pixel_sum)
```

d. Indexul imaginii cu suma maxima

```python
max_index = np.argmax(pixel_sum)
print('max index =', max_index)
```
